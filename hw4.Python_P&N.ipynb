__*Задание 3__

С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3 включительно.

Создайте массивы y1, y2, y3, y4 по следующим формулам:

y1 = x**2

y2 = 2 * x + 0.5

y3 = -3 * x - 1.5

y4 = sin(x)

Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.

Для графика в левом верхнем углу установите границы по оси x от -5 до 5.

Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.

Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.


```python
x = np.linspace(-3, 3, 51)

y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title('График y1', color = 'Green')
ax1.set_xlabel('x', loc = 'right')
ax1.set_ylabel('y', loc = 'top')
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2', color = 'Red')
ax2.set_xlabel('x', loc = 'right')
ax2.set_ylabel('y', loc = 'top')

ax3.plot(x, y3) 
ax3.set_title('График y3', color = 'Blue')
ax3.set_xlabel('x', loc = 'right')
ax3.set_ylabel('y', loc = 'top')

ax4.plot(x, y4)
ax4.set_title('График y4', color = 'Orange')
ax4.set_xlabel('x', loc = 'right')
ax4.set_ylabel('y', loc = 'top')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

На следующем графике постройте две гистограммы по значениям признака V1 - одну для мошеннических транзакций (Class равен 1) и другую - для обычных (Class равен 0). Число бинов должно равняться 20 для обеих гистограмм, а коэффициент alpha сделайте равным 0.5, чтобы гистограммы были полупрозрачными и не загораживали друг друга. Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “V1”.


```python
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
creditcard = pd.read_csv('creditcard.csv')
```

```python
creditcard.head(10)
```

```python
creditcard.info()
```

```python
# показывает, что в столбце Class есть только значения 0 и 1, кот-ые повторяются n-раз соот-в-но
class_list = creditcard['Class'].value_counts()
print(class_list)
```

```python
class_list.plot(kind = 'barh')
plt.show()
```

```python
class_list.plot(kind = 'barh', logx = True)
plt.show()
```

```python
#  с пом метода loc идет выборка знач-й: в данном случае 2 столбца, кот-е пересекаются
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
```

```python
#alpha - прозрачность
#bins - кол-во бинов(столбцов) в гисттограмме
#label - данные в левом верхнем углу (легенда-задаем заранее в переменных, потом просто вызываем отдельно)
#density -
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.xlabel('V1', loc = 'right')
plt.legend()
plt.show()
```
